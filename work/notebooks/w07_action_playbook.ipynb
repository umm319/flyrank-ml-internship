## Intended Use

This action playbook is designed to help prioritize SEO and content improvement tasks. It supports decision-making by identifying pages that may benefit from updates, CTR optimization, keyword improvements, and better internal linking. The recommendations are intended to assist content teams and should always be reviewed before implementation.

## Limits

- The recommendations are based on the available dataset and selected signals.
- The model does not guarantee better rankings or traffic.
- Human review is required before making content changes.
- Results may vary for different websites or datasets.
- This playbook is intended as decision support and should not be used for fully automated publishing.

## Human Review

Before applying any recommendation, a human reviewer should verify:

- The page content is accurate and up to date.
- The recommendation matches the user's search intent.
- No important information is removed during optimization.
- The suggested changes align with business goals and SEO best practices.
- The recommendation is supported by the available data.

## No-Go List

The following actions should **NOT** be automated:

- Publishing or deleting content without human approval.
- Making medical, legal, or financial content changes automatically.
- Changing factual information without verification.
- Automatically rewriting pages with low confidence.
- Applying recommendations when the available data is incomplete or unreliable.

## Monitoring

The model should be monitored regularly to ensure that its recommendations remain useful and accurate. Performance metrics such as CTR, impressions, and ranking positions should be reviewed over time.

## Retrain Triggers

The model should be retrained when:

- Website traffic patterns change significantly.
- New content or pages are added.
- Model performance decreases over time.
- Data distribution changes (data drift).
- SEO strategies or business goals are updated.

Regular monitoring and retraining help maintain reliable and trustworthy recommendations.
